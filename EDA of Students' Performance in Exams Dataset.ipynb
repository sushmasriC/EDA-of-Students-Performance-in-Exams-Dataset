```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Understand the Dataset
# Load the dataset into a DataFrame
df = pd.read_csv("Students.csv")  # Ensure the dataset is in the same directory as this script.

# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Show the structure of the dataset (number of rows, columns, and data types)
print("\nStructure of the dataset:")
print(df.info())

# Provide a statistical summary of numerical columns
print("\nSummary of numerical data:")
print(df.describe())

# Task 2: Descriptive Statistics
# Calculate the mean, median, standard deviation, minimum, and maximum for math, reading, and writing scores
subjects = ['math score', 'reading score', 'writing score']

for subject in subjects:
    print(f"\n{subject.capitalize()}:")
    print(f"Mean: {df[subject].mean():.2f}")
    print(f"Median: {df[subject].median():.2f}")
    print(f"Standard Deviation: {df[subject].std():.2f}")
    print(f"Min: {df[subject].min()}")
    print(f"Max: {df[subject].max()}")

# Task 3: Data Visualizations

# Create histograms for the distribution of scores in math, reading, and writing
plt.figure(figsize=(12, 6))
for i, subject in enumerate(subjects, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[subject], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution of {subject.capitalize()}')
    plt.xlabel(subject.capitalize())
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Create a bar chart showing average scores in each subject categorized by gender
gender_scores = df.groupby('gender')[subjects].mean().reset_index()
gender_scores.set_index('gender').plot(kind='bar', figsize=(8, 5), color=['#ff9999', '#66b3ff', '#99ff99'])
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xlabel('Gender')
plt.legend(title='Subjects')
plt.show()

# Create a scatter plot to show the relationship between math and reading scores
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='math score', y='reading score', hue='gender', palette='coolwarm')
plt.title('Math vs Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(title='Gender')
plt.show()

# Task 4: Categorical Data Analysis

# Analyze the effect of parental education level on average scores
edu_scores = df.groupby('parental level of education')[subjects].mean().reset_index()
edu_scores.plot(x='parental level of education', kind='bar', figsize=(10, 6), color=['#ffcccb', '#87cefa', '#90ee90'])
plt.title('Average Scores by Parental Education Level')
plt.ylabel('Average Score')
plt.xlabel('Parental Level of Education')
plt.xticks(rotation=45)
plt.legend(title='Subjects')
plt.show()

# Analyze the effect of test preparation on average scores
prep_scores = df.groupby('test preparation course')[subjects].mean().reset_index()
prep_scores.set_index('test preparation course').plot(kind='bar', figsize=(8, 5), color=['#ff9999', '#66b3ff', '#99ff99'])
plt.title('Effect of Test Preparation on Scores')
plt.ylabel('Average Score')
plt.xlabel('Test Preparation Course')
plt.legend(title='Subjects')
plt.show()

# Task 5: Insights
# Print insights based on visualizations and statistics
insights = """
Insights:
1. Gender: Females tend to perform better in reading and writing, while males show a slight edge in math.
2. Parental Education: Higher levels of parental education correlate with better performance in all subjects.
3. Test Preparation: Students who completed the test preparation course scored significantly higher in all subjects.
"""
print(insights)
```
